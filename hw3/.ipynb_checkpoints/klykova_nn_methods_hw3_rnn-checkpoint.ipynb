#### Dataset & Dataloader

```python
class MyDataset(Dataset):

    def __init__(self, dataset, word2id, max_len, DEVICE):
        self.dataset = dataset['tokens'].values
        self.target = dataset['target'].values
        self.length = dataset.shape[0]
        self.word2id = word2id
        self.max_len = max_len
        self.device = DEVICE

    def __len__(self):
        return self.length

    def __getitem__(self, index):
        words = self.dataset[index]
        # паддинг и обрезка по maxlen
        ids = torch.LongTensor([self.word2id[word] if word in self.word2id
                                else self.word2id['UNK'] for word in words]
                               [:self.max_len])
        y = [self.target[index]]
        return ids, y

    def collate_fn(self, batch):
        ids, y = list(zip(*batch))
        padded_ids = torch.vstack(
            [F.pad(idx, pad=(0, self.max_len - idx.shape[0]),
                   mode='constant', value=0)
             for idx in ids])
        padded_ids = pad_sequence(ids, batch_first=True).to(self.device)
        y = torch.LongTensor(y).T[0].to(self.device)
        return padded_ids, y
```

```python
train_dataset = MyDataset(train_data, word2id, MAX_LEN, DEVICE)
train_sampler = RandomSampler(train_dataset)
train_iterator = DataLoader(train_dataset,
                            collate_fn=train_dataset.collate_fn,
                            sampler=train_sampler, batch_size=128)
```

```python
val_dataset = MyDataset(val_data, word2id, MAX_LEN, DEVICE)
val_sampler = SequentialSampler(val_dataset)
val_iterator = DataLoader(val_dataset,
                          collate_fn=val_dataset.collate_fn,
                          sampler=val_sampler, batch_size=128)
```

#### Класс модели

```python
class C_LSTM(nn.Module):

    def __init__(self, weights, vocab_size, embedding_dim=300,
                 max_length=512, output_dim=6,
                 filter_sizes=[2, 3, 4], num_filters=150,
                 lstm_layers=1, memory_dim=150,
                 dropout_input=False, dropout_lstm=True,
                 dropout_rate=0.5):

        super().__init__()

        self.vocab_size = vocab_size
        self.max_length = max_length
        self.filter_sizes = filter_sizes
        self.cnn_layers = len(filter_sizes)

        # cut the window sequence based on the maximum filter length
        self.max_features = max_length - max(filter_sizes) + 1

        self.embedding = nn.Embedding.from_pretrained(weights)

        self.dropout_input = dropout_input
        self.dropout_lstm = dropout_lstm
        self.dropout = nn.Dropout(p=dropout_rate)

        conv_layers = []
        for f_size in filter_sizes:
            conv = nn.Conv1d(in_channels=embedding_dim,
                             out_channels=num_filters,
                             kernel_size=f_size, padding='valid')
            conv_layers.append(conv)
        self.conv_layers = nn.ModuleList(conv_layers)

        self.lstm = nn.LSTM(input_size=len(filter_sizes)*num_filters,
                            hidden_size=memory_dim, num_layers=lstm_layers,
                            batch_first=True)

        self.relu = nn.ReLU()
        self.hidden = nn.Linear(in_features=memory_dim,
                                out_features=output_dim)
        self.out = nn.Softmax(dim=1)

    def forward(self, word):

        # эмбеддинги и дропаут
        embedded = self.embedding(word)
        embedded = embedded.transpose(1, 2)
        if self.dropout_input:
            embedded = self.dropout(embedded)

        # CNN
        feature_maps = []
        for conv in self.conv_layers:
            feature_map = self.relu(conv(embedded))[:, :, :self.max_features]
            feature_maps.append(feature_map)

        if self.cnn_layers > 1:
            rnn_input = torch.cat(feature_maps, 1).transpose(1, 2)
        else:
            rnn_input = feature_map.transpose(1, 2)

        # LSTM
        _, (hidden_state, _) = self.lstm(rnn_input)
        if self.dropout_lstm:
            embedded = self.dropout(hidden_state)
        logits = self.out(self.hidden(torch.squeeze(hidden_state, 0)))

        return logits
```

```python
def train(model, iterator, optimizer, criterion, n_epoch, metric):

    progress_bar = tqdm(total=len(train_iterator.dataset),
                        desc='Epoch {}'.format(n_epoch + 1))

    epoch_losses = []  # для подсчета среднего лосса на всех батчах
    epoch_metrics = []
    model.train()  # ставим модель в обучение

    for i, (texts, ys) in enumerate(iterator):  # итерируемся по батчам

        optimizer.zero_grad()  # обнуляем градиенты
        preds = model(texts)  # прогоняем данные через модель
        loss = criterion(preds, ys)  # считаем значение функции потерь
        loss.backward()  # считаем градиенты
        optimizer.step()  # обновляем веса

        # сохраняем значения лосс и метрику
        epoch_losses.append(loss.item())
        batch_metric = metric(preds.argmax(1).long(), ys.long(),
                              average='weighted', num_classes=6,
                              ignore_index=0)
        epoch_metrics.append(batch_metric.cpu().numpy())

        if (i + 1) % int(len(iterator) / 5) == 0:
            print('Train loss: {}, Train metric: {}'.format(
                np.mean(epoch_losses), np.mean(epoch_metrics)))

        progress_bar.update(texts.shape[0])

    progress_bar.close()

    # возвращаем список (потом построим график и посчитаем среднее)
    return epoch_losses, epoch_metrics
```

```python
def evaluate(model, iterator, criterion, n_epoch, metric):

    epoch_losses = []
    epoch_metrics = []
    model.eval()

    with torch.no_grad():

        for i, (texts, ys) in enumerate(iterator):
            preds = model(texts)  # делаем предсказания на тесте
            loss = criterion(preds, ys)  # считаем значения лосса

            # сохраняем значения лосса и метрику
            epoch_losses.append(loss.item())
            batch_metric = metric(preds.argmax(1).long(), ys.long(),
                                  average='weighted', num_classes=6,
                                  ignore_index=0)
            epoch_metrics.append(batch_metric.cpu().numpy())

            if (i + 1) % int(len(iterator) / 5) == 0:
                print('Val loss: {}, Val metric: {}'.format(
                    np.mean(epoch_losses), np.mean(epoch_metrics)))

    # возвращаем список (потом построим график и посчитаем среднее)
    return epoch_losses, epoch_metrics
```

```python
def train_and_evaluate(model, train_iterator, val_iterator,
                       optimizer, criterion, metric, epochs=5):
    losses_train = []
    metrics_train = []
    losses_eval = []
    metrics_eval = []

    for n_epoch in range(epochs):

        print('\nTraining...')
        train_losses, train_metrics = train(model, train_iterator,
                                            optimizer, criterion,
                                            n_epoch, metric)
        losses_train.append(train_losses)
        metrics_train.append(train_metrics)

        print('\nValidating...\n')
        test_losses, test_metrics = evaluate(model, val_iterator,
                                             criterion, n_epoch,
                                             metric)
        losses_eval.append(test_losses)
        metrics_eval.append(test_metrics)

    return losses_train, losses_eval, metrics_train, metrics_eval
```

```python
def plot_train_and_eval(losses_train, losses_eval,
                        metrics_train, metrics_eval,
                        params):

    all_train_losses = []
    all_train_metrics = []
    for epoch, losses in enumerate(losses_train):
        all_train_losses.extend(losses)
        all_train_metrics.extend(metrics_train[epoch])

    all_eval_losses = []
    all_eval_metrics = []
    for epoch, losses in enumerate(losses_eval):
        all_eval_losses.extend(losses)
        all_eval_metrics.extend(metrics_eval[epoch])

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    fig.suptitle(params)

    ax1.set_title('Training Process')
    ax1.set_xlabel('Iterations')
    ax1.set_ylabel('Metric')
    ax1.plot(all_train_losses, label='loss')
    ax1.plot(all_train_metrics, label='accuracy')
    ax1.legend()
    ax1.grid()

    ax2.set_title('Validation')
    ax2.set_xlabel('Iterations')
    ax2.set_ylabel('Metric')
    ax2.plot(all_eval_losses, label='loss')
    ax2.plot(all_eval_metrics, label='accuracy')
    ax2.legend()
    ax2.grid()
```

```python
plot_train_and_eval(losses_train, losses_eval, metrics_train, metrics_eval)
```

```python
def plot_loss_and_acc(losses_train, losses_eval,
                      metrics_train, metrics_eval,
                      params):

    mean_train_losses = []
    mean_train_metrics = []
    for epoch, losses in enumerate(losses_train):
        mean_train_losses.append(np.mean(losses))
        mean_train_metrics.append(np.mean(metrics_train[epoch]))

    mean_eval_losses = []
    mean_eval_metrics = []
    for epoch, losses in enumerate(losses_eval):
        mean_eval_losses.append(np.mean(losses))
        mean_eval_metrics.append(np.mean(metrics_eval[epoch]))

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    fig.suptitle(params)

    ax1.set_title('Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Metric')
    ax1.plot(mean_train_losses, label='train')
    ax1.plot(mean_eval_losses, label='test')
    ax1.legend()
    ax1.grid()

    ax2.set_title('Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Metric')
    ax2.plot(mean_train_metrics, label='train')
    ax2.plot(mean_eval_metrics, label='test')
    ax2.legend()
    ax2.grid()
```

```python
def parameter_search(weights, word2id, DEVICE):
    results = []
    # берем разное число и размер фильтров, как в статье
    filter_types = [[2], [3], [4], [2, 3], [3, 4], [2, 4], [2, 3, 4]]

    for f_type in filter_types:
        c_lstm = C_LSTM(weights=weights, vocab_size=len(word2id),
                        filter_sizes=f_type)
        optimizer = torch.optim.RMSprop(c_lstm.parameters(), lr=0.001,
                                        weight_decay=0.0001)
        criterion = nn.CrossEntropyLoss()
        c_lstm = c_lstm.to(DEVICE)
        criterion = criterion.to(DEVICE)

        print('\n\n')
        print('Filters:', f_type)

        losses_train, losses_eval, metrics_train, metrics_eval = \
            train_and_evaluate(c_lstm, train_iterator, val_iterator,
                               optimizer, criterion, accuracy, epochs=5)

        results.append({'filter_types': f_type,
                        'train_losses': losses_train,
                        'train_accs': metrics_train,
                        'val_losses': losses_eval,
                        'val_accs': metrics_eval})
    return results
```

```python
results = parameter_search(weights, word2id, DEVICE)
```
